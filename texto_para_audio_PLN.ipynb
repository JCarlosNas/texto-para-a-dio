```python
from gtts import gTTS
```

```python
text_to_say = "My favorite fruit is passion fruit"

language = "en"

gtts_object = gTTS(text = text_to_say, lang = language, slow = False)

gtts_object.save("/content/gtts.wav")



```

```python
from IPython.display import Audio

Audio("/content/gtts.wav")
```

```python
french_text = "Je vais au supermarché aujourd'hui "

french_language = "fr"

french_gtts_object = gTTS(text = french_text, lang = french_language, slow = True)

french_gtts_object.save("/content/french.wav")

Audio("/content/french.wav")
```

```python
portu_text = "Final de ano"

portu_language = "pt"

portu_gtts_object = gTTS(text = portu_text, lang = portu_language, slow= False)

portu_gtts_object.save("/content/portu.wav")

Audio ("/content/portu.wav")
```
